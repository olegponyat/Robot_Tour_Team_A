```python
import numpy as np
import matplotlib.pyplot as plt

%store -r distance offset

def compute_circular_path(d_t, d_o, num_points=100000):
    theta = 4 * np.arctan(2 * d_o / d_t)

    r = (d_t / 2) / np.sin(theta / 2)

    angle_vals = np.linspace(-theta / 2, theta / 2, num_points)
    x = r * np.cos(angle_vals)
    y = r * np.sin(angle_vals)

    return theta, r, x, y

def plot_full_circular_path(x, y, r):
    plt.figure(figsize=(8, 8))

    plt.plot(x, y, 'b', label='Arc Path')

    plt.scatter([x[0], x[-1]], [y[0], y[-1]], color='green', label='Start/End')

    plt.scatter(0, 0, color='red', label='Center (O)')

    plt.plot([x[0], x[-1]], [y[0], y[-1]], 'm', label='Straight Distance')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.axis('equal')
    plt.legend()
    plt.title('Circular Path of the Car')
    plt.grid(True)
    plt.show()

theta, r, x, y = compute_circular_path(distance, offset)

arc_length = r * theta

print(f"central angle: {np.degrees(theta)}")
print(f"arc length: {arc_length}")

plot_full_circular_path(x, y, r)
```
